```python
# importing the module
import pandas as pd
# creating a sample dataframe
data = pd.DataFrame({'Brand' : ['Maruti', 'Hyundai', 'Tata',
								'Mahindra', 'Maruti', 'Hyundai',
								'Renault', 'Tata', 'Maruti'],
					'Year' : [2012, 2014, 2011, 2015, 2012,
							2016, 2014, 2018, 2019],
					'Kms Driven' : [50000, 30000, 60000,
									25000, 10000, 46000,
									31000, 15000, 12000],
					'City' : ['Gurgaon', 'Delhi', 'Mumbai',
							'Delhi', 'Mumbai', 'Delhi',
							'Mumbai','Chennai', 'Ghaziabad'],
					'Mileage' : [28, 27, 25, 26, 28,
								29, 24, 21, 24]})

# displaying the DataFrame
display(data)

# saving the dataframe
data.to_csv('file1.csv')
```

loc() : loc() is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc() . Many operations can be performed using the loc() method like-

```python
# selecting cars with brand 'Maruti' and Mileage > 25
# Excel multiple field filtering
data.loc[(data.Brand == 'Maruti') & (data.Mileage > 25)]
```

```python
# selecting range of rows from 2 to 5
data.loc[2:5] #ending is NOT (ending -1) when using loc
```

```python
# updating values of Mileage if Year < 2015
# First filter based on the year then for years less than 2015 change the mileage to 22
data.loc[(data.Year <2015), ['Mileage']] = 22
display(data)
```

iloc() : iloc() is a indexed based selecting method which means that we have to pass integer index in the method to select specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc(). Operations performed using iloc() are:

```python
# selecting 0th, 2th, 4th, and 7th index rows
# data.iloc[list]
# data.iloc[ [1,2,3] ]
data.iloc[[0,2,4,7]] #[[]] - matrix which is m x n
```

```python
# selecting rows from 1 to 4 and columns from 2 to 4
data.iloc[1:5, 2:5] #the logic of (end-1) returns with iloc
```

```python

# importing pandas
import pandas as pd
  
record = {
  'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
  'Age': [21, 19, 20, 18, 17, 21],
  'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
  'Percentage': [88, 92, 95, 70, 65, 78]}
  
# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])
  
print("Given Dataframe :\n", dataframe) 
  
# selecting rows based on condition
rslt_df = dataframe.loc[dataframe['Percentage'] > 80]
  
print('\nResult dataframe :\n', rslt_df)
```

Vertical and horizontal concatenation of tables

```python
import numpy as np
import pandas as pd
df1 = pd.DataFrame(np.random.randint(25, size=(4, 4)),
                   index=["1", "2", "3", "4"],
                   columns=["A", "B", "C", "D"])
 
df2 = pd.DataFrame(np.random.randint(25, size=(6, 4)),
                   index=["5", "6", "7", "8", "9", "10"],
                   columns=["A", "B", "C", "D"])
 
df3 = pd.DataFrame(np.random.randint(25, size=(4, 4)),
                   columns=["A", "B", "C", "D"])
 
df4 = pd.DataFrame(np.random.randint(25, size=(4, 4)),
                   columns=["E", "F", "G", "H"])
 
display(df1, df2, df3, df4)
```

```python
# concatenating df1 and df2 along rows
vertical_concat = pd.concat([df1,df2], axis = 0)

#axis = 0 is for y-axis 
#same number of columns in both tables
#Axis = 1 is for x-axis

# concatenating df3 and df4 along columns
horizontal_concat = pd.concat([df3,df4], axis = 1)

display(vertical_concat, horizontal_concat)
```
