```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y, 'r') #b stands for blue, r is red, y is yellow, c is cyan, g is green, k is black, s is blue dotted line
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
#plt.subplot(nrow, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
#plt.subplot(nrow, ncols, plot_number)
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.subplot(2,1,1)
plt.plot(y,x,'g*-')
```

```python
#plt.subplot(nrow, ncols, plot_number)
plt.subplot(2,3,1)
plt.plot(x,y,'r--')
plt.subplot(2,3,6)
plt.plot(y,x,'g*-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height

axes.plot(x,y,'b')
axes.set_xlabel("Set X Label")
axes.set_ylabel("Set Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height
axes2 = fig.add_axes([0.2,0.5,0.4,0.6]) #left, bottom, width, height

axes1.plot(x,y,'b')
axes1.set_xlabel("Set X1 Label")
axes1.set_ylabel("Set Y1 Label")
axes1.set_title("Set Title1")

axes2.plot(x,y,'r')
axes2.set_xlabel("Set X2 Label")
axes2.set_ylabel("Set Y2 Label")
axes2.set_title("Set Title2")
```
