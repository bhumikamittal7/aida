```python
#===============================================================================
#=================importing libraries==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#=================ignore warnings==========================================
import warnings
warnings.filterwarnings(action='once')
warnings.filterwarnings('ignore')
```

Plot the graph of y = x^3

```python
#===============================================================================
#=================ploting the graph ============================================
x = np.linspace(0,5,11)
y = x**3

plt.plot(x,y, 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^3 graph \n')
plt.show()
```

Create the given dataframe using list.

```python
# creating the dataframe
lst = [['Ram', 40], ['Tom', 30], ['Shyam', 50], ['Rachel',70]]
data = pd.DataFrame(lst, columns=['Name', 'Marks'])

# displaying the DataFrame
data
```

In the “Churn Modelling” dataset provided in class apply filtering and determine how many ‘male’ customers are in the bank AND hail from Germany.

```python
#=================read file and filter =========================================
data = pd.read_csv("Churn_Modelling.csv")
df = data[data['Gender'].str.contains('Male')]
df1 = df[df['Geography'].str.contains('Germany')]

print("Number of ‘male’ customers from 'Germany' = ", len(df1.index), "\n")
df1
```

Create a histogram showing how many customers are from which country.

```python
CountryData = data.groupby(['Geography'])[['Geography']].count()
lst = list(CountryData.index)
display(CountryData)

data_dict = CountryData.to_dict()
df = data_dict['Geography']

country = list(df.keys())
density = list(df.values())

plt.bar(country, density, width = 0.3);

plt.xlabel('Country')
plt.ylabel('Density')
plt.title('Geography \n')
plt.show()
```

Create a pie-chart showing how many customers are from which country.

```python
CountryData = data.groupby(['Geography'])[['Geography']].count()
lst = list(CountryData.index)
display(CountryData)

plt.pie(CountryData, labels = lst);
```

Make a histogram of customers exited/non exited

```python
ExitedData = data.groupby(['Exited'])[['Exited']].count()
display(ExitedData)
lst = list(ExitedData.index)

data_dict = ExitedData.to_dict()
df = data_dict['Exited']

exited = list(df.keys())
density = list(df.values())

plt.bar(exited, density, width = 0.1);

plt.xlabel('Exited?')
plt.ylabel('Density')
plt.show()
```
