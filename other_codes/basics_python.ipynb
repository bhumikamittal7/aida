2.1 Create a dataframe using dictionary in Python for the following data. The data frame should look exactly as shown below. Note the first and last name have been separated by a blank.

```python
# creating the dataframe
data = pd.DataFrame({'Name' : ['ram ji', 'santa banta', 'peppa pig',
								'santa claus', 'pepsi coke'],
					'Marks' : [80,30,70,20,60]})

# displaying the DataFrame
data
```

2.2 Write Python code to assign students who have got less than 60 marks the grade “Fail”. Rest should be assigned “Pass”. This assignment should be in a new column whose column header is “Grade”

```python
data['Grade'] =[0]*5
for i in data.index:
  if (data['Marks'][i]< 60):
    data['Grade'][i] = 'Fail'
  else:
    data['Grade'][i] = 'Pass'

data
```

2.3 Separate the first name and last name in the data frame in two separate columns. The new two columns should read as First Name and Last Name

```python
data[['First Name','Last Name']] = data.Name.str.split(expand=True)
data.drop('Name', axis=1, inplace=True)

data.insert(0, 'First Name', data.pop('First Name'))
data.insert(1, 'Last Name', data.pop('Last Name'))

data
```

2.4 Ensure that only last names are CAPITALIZED

```python
for i in data.index:
  data['Last Name'][i] = data['Last Name'][i].upper()

data
```

2.5 After creating the dataframe convert it to a csv file by the name “Data” and download it.Before downloading, check that the following columns have been created First Name, Last Name, Marks, Grade

```python
data.to_csv('data.csv')
```

2.6 Make a bar chart with the x-axis being the name and y-axis being the marks. The bar chart should have X label and Y label. It should have a title.

```python
name = [0]*5

for i in data.index:
  name[i] = data['First Name'][i] + " " + data['Last Name'][i].lower()

marks = list(data['Marks'])

plt.bar(name, marks, width = 0.3);

plt.xlabel('Name')
plt.ylabel('Marks')
plt.title('Mark-wise Distribution \n')
plt.show()
```

2.7 Make a pie-chart showing percentage failed and passed. Label the chart properly.

```python
grade = data.groupby(['Grade'])[['Grade']].count()
lst = list(grade.index)
plt.pie(grade, labels = lst, autopct='%1.1f%%');

plt.title('Grade Distribution \n')
plt.show()
```

2.8 Write a for loop to count the number of students who are in the dataframe. The output expected is ‘Number of students are: 5’

```python
count = 0
for i in data.index:
  while True:
    count += 1
    break

print('Number of students are: {}'.format(count))
```

2.9 Using Python determine what is the average marks scored in the exam?

```python
lst = []
for i in data.index:
  lst.append(data['Marks'][i])

avg = sum(lst) / len(lst)
print('The average marks scored in the exam: {}'.format(avg))
```

2.10 Using Python determine what percentage of students have passed the exam?

```python
lst = []
for i in data.index:
  if (data['Grade'][i] == 'Pass'):
    lst.append(1)
  else:
    lst.append(0)

ratio = sum(lst)/len(lst) 
percent = ratio*100
percent = int(round(percent))

print('The percentage of students have passed the exam: {}%'.format(percent))
```

3.3 Square each number in the list and call it x_squared

```python
x_squared = [x**2 for x in X]
print(x_squared)
```

3.4 Cube each number in the list and call it x_cube

```python
x_cube = [x**3 for x in X]
print(x_cube)
```

3.5 Create a list where every element in X_cube subtracts the corresponding element in X_square. Call this list Y

```python
Y = [0]*10
for i in range(10):
  Y[i] = x_cube[i] - x_squared[i]

print(Y)
```

3.6 Square every element in Y and sum it up. The output should be displayed as a number

```python
y_squared = [y**2 for y in Y]

sumY = sum(y_squared)
print(sumY)
```
